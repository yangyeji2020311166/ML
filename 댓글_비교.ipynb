```python
f=open('/content/스크립트.txt','r',encoding='utf-8')
txt = f.readlines()
f.close()
```

```python
from konlpy.tag import Okt
twitter = Okt()
morphs = []

for sentence in txt:
    morphs.append(twitter.pos(sentence))
    
print(morphs)
```

```python
noun_list = []

for i in morphs:
    for word, tag in i:
        if tag in['Noun']:
            noun_list.append(word)

print(noun_list)
```

```python
f = open("/content/stopwords_korean.txt", "rt", encoding="utf-8")
lines = f.readlines()
stopwords = []
for line in lines:
    line = line.replace('\n', '')
    stopwords.append(line)
f.close()

unique_Noun_words = set(noun_list)
for word in unique_Noun_words:
    if word in stopwords:
        while word in noun_list: noun_list.remove(word)
```

```python
print(unique_Noun_words)
```

```python
from collections import Counter
count = Counter(noun_list)

words = (dict(count.most_common()))
print(words)
```
